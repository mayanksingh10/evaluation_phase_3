# birvariate analysis

```python
def stacked_plot(df, group, target):
    fig, ax = plt.subplots(figsize = (6,4))
    temp_df = (df.groupby([group, target]).size()/df.groupby(group)[target].count()).reset_index().pivot(columns=target, index=group, values=0)
    temp_df.plot(kind='bar', stacked=True, ax = ax, color = ["green", "red"])
    ax.xaxis.set_tick_params(rotation=0)
    ax.set_xlabel(group)
    ax.set_ylabel('Churn Percentage')
```

```python
from sklearn.model_selection import train_test_split
```

```python
maxacc=0
maxrs=0
for i in range(1,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    lr=LogisticRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxacc:
        maxacc=acc
        maxrs=i

print("r2 score=",maxacc,"random_state",maxrs)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=57)
```

```python
lr = LogisticRegression()
gb= GaussianNB()
dtc= DecisionTreeClassifier()
svc=SVC()
rf=RandomForestClassifier()
et=ExtraTreesClassifier()
ada=AdaBoostClassifier()
gbc=GradientBoostingClassifier()
models=[lr,gb,dtc,svc,rf,et,ada,gbc]
for i in models:
    print(i)
    i.fit(x_train,y_train)
    pred=i.predict(x_test)
    print(accuracy_score(y_test,pred))
    print(confusion_matrix(y_test,pred))
    print(classification_report(y_test,pred))
```
